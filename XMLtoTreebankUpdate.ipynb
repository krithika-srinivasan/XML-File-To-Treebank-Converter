# Parser

```python
# Parser rules:
#     < becomes (
#     > becomes ""
#     fct= becomes -
#     "SUJ" becomes SUJ
#     </ becomes )
    
```

```python
def getNext(i):
    return inputchar[i+1]

def getPrev(i):
    return inputchar[i-1]

def replaceEndTag(c): #i is the index of the '/'
    outputchar.pop() #Replace the '(' with a blank
    outputchar.insert(c,')')
```

```python
c = 0
outputchar = []
i = 0

while(i < len(inputchar)):

    if(inputchar[i] == '<'):
        outputchar.insert(c,'(')
        c = c + 1
        i = i + 1
    
    #Clean up sent
    elif(inputchar[i] == 'a' and getNext(i) == 'r'  and getNext(i+1) == 'g'and getNext(i+2) == 'u'and getNext(i+3) == 'm'
         and getNext(i+4)== 'e' and getNext(i+5) == 'n' and getNext(i+6) == 't'and getNext(i+7) == '='):
        outputchar.insert(c,'')
        c = c + 1
        while(inputchar[i] != '>'):
            i = i + 1
    #Erase ei -> lemma
    elif(inputchar[i] == 'e' and getNext(i) == 'i'):
        outputchar.insert(c,'')
        c = c + 1
        while(inputchar[i] != 'l'):
            i = i + 1
    
    #Erase mph to end of word tag>
    elif(inputchar[i] == 'm' and getNext(i) == 'p'  and getNext(i+1) == 'h'):
        outputchar.insert(c,'@')
        c = c + 1
        while(inputchar[i] != '>'):
            i = i + 1
    

            
    elif(inputchar[i] == '>'):
        outputchar.insert(c,'@')
        c = c + 1
        i = i + 1


    elif(inputchar[i] == '/' and getPrev(i) == '<'):
        replaceEndTag(c)
        c = c + 1
        while(inputchar[i] != '>'):
            i = i + 1

#Base Case
    else:
        outputchar.insert(c,inputchar[i])
        c = c + 1
        i = i + 1
```
